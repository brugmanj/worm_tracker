```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set()
```

## Plotting Accelerations per Mutation

```python
filepath = '../../data/accelerations/'

files = os.listdir(filepath)

legend = []

for file in files:
    df = pd.read_csv(filepath+file)
    df = df.mean(axis = 1)
    legend.append(file.strip('.csv'))
    plt.plot(df)
    plt.title('MUTATNT')
    plt.xlabel('Frames')
    plt.ylabel('Pixels/Frame^2')
    plt.savefig('../../plots/' + file.strip('.csv') + '.png', dpi = 300)
    plt.show()
```

## Plotting 3 Mutations

```python
files = ['REDACTED']

for file in files:
    df = pd.read_csv(filepath+file)
    df = df.mean(axis = 1)
    legend.append(file.strip('.csv'))
    plt.plot(df)
    plt.title('Comparing Mutation Acceleration')
    plt.xlabel('Frames')
    plt.ylabel('Pixels/Frame^2')
    plt.legend(['MUTANT', 'MUTANT', 'MUTANT'], title = 'Mutations')
    plt.savefig('../../plots/' + 'MUTANTS' + '.png', dpi = 300)
```

```python
more_files = os.listdir('../../data/frame_distances/REDACTED/')
df = []
for f in more_files:
    df.append(pd.read_csv('../../data/frame_distances/REDACTED/' + f))

df = pd.concat(df, axis = 1)
df = df.mask(df > 9)

plt.figure(figsize = (12, 8))
sns.distplot(df.mean(axis = 1), )
plt.xlabel('Speed (pixels/frame)')
plt.title('Speed Distribution in worm', size = 20)
plt.savefig('../../plots/speedworm.png', dpi = 300)
```
